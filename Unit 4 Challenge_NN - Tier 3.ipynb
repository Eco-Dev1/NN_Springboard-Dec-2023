```python
#create a function that will divide the average price of a borough in 1998 by the averae price of the same
#borough in 2018
def ratio_1998_2018(i):
    y1998=(i["average_price"].values[i["year"]==1998])
    y2018=(i["average_price"].values[i["year"]==2018])
    ratio= y1998/y2018
    return ratio

#test the function
ratio_1998_2018(df[df["London_Borough"]=="Barking & Dagenham"])
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
# create a function to compare the absolute change in average price of a borough in 1998 compared to 2018 
def diff_1998_2018(i):
    y1998=(i["average_price"].values[i["year"]==1998])
    y2018=(i["average_price"].values[i["year"]==2018])
    diff= y2018-y1998
    return diff

#test the function
diff_1998_2018(df[df["London_Borough"]=="Barking & Dagenham"])
```

```python
# make empty dictionary to accumulate and store ratios of borough prices
ratio_dic ={}
# use iterator "b" to go through all the boroughs and enter the borough and associated price ratio in the dictionary. ratio_dic
for b in df["London_Borough"].unique():
    borough=df[df["London_Borough"]==b]
    ratio_dic[b]=ratio_1998_2018(borough)
    
print(ratio_dic)
```

```python
df_ratios=pd.DataFrame(ratio_dic)
print(df_ratios.head(15))
```

```python
# make empty dictionary to accumulate and store differences of borough prices
diff_dic ={}
# use iterator "b" to go through all the boroughs and enter the borough and associated price ratio in the dictionary. ratio_dic
for b in df["London_Borough"].unique():
    borough=df[df["London_Borough"]==b]
    diff_dic[b]=diff_1998_2018(borough)
    
df_diff=pd.DataFrame(diff_dic)
print(df_diff.head(15))
```

```python
# transpose the dataframes and reset the indices,
ratio_table=df_ratios.T.reset_index()
diff_table=df_diff.T.reset_index()
print(ratio_table.head())
print(diff_table.head())

```

```python
# Rename columns for vizualization
ratio_table.rename(columns={"index":"Borough",0:"ratio(1998/2018)"},inplace=True)
diff_table.rename(columns={"index":"Borough",0:"diff(2018-1998)"},inplace=True)
print(ratio_table.head())
print(diff_table.head())
```

```python
#Greatest price differences will have the lowest ratios. Sort in ascending order
sort=ratio_table.sort_values("ratio(1998/2018)")
sort.plot(kind="bar", x="Borough", y="ratio(1998/2018)")
plt.show()
```

```python
#show greatest absolute differences
sort1=diff_table.sort_values("diff(2018-1998)", ascending=False)
sort1.plot(kind="bar", x="Borough", y="diff(2018-1998)")
plt.show()
```
